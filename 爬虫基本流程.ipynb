设置请求头，模拟浏览器发送请求

```python
import requests
url='http://ligong.sanyau.edu.cn'
headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0'}   # 设置请求头,模拟浏览器发送请求
rq = requests.get(url, headers=headers)
rq.headers     # 返回的响应头
```

查看响应内容

```python
print('响应状态码：', rq.status_code)
print('编码：', rq.encoding)
print('请求头：', rq.headers)
print('实体：', rq.text)    # rq.content与rq.text类似，但返回值是bytes类型
```

生成完整的HTTP请求


```python
import requests
url='http://ligong.sanyau.edu.cn'     # 目标网址
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
rq = requests.get(url, headers=headers) # 设置请求头,模拟浏览器发送请求
rq.encoding = rq.apparent_encoding    # 设置编码方式
print(rq.text)      #获取相应内容
```

使用Beautiful Soup解析网页


```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(rq.text,'lxml')  #用lxml解析器进行解析


```

定位目标内容：获取某些标签内容


```python
soup.head     # 获取网页的head标签
```

```python
soup.body     # 获取网页的body标签
```

```python
soup.a           # 获取网页的第一个a标签
```

定位目标内容：CSS Selector


```python
soup.select('title')   # 获取soup对象中的所有title标签
```

基本selector写法


```python
soup.select('p')     #查找所有p标签
```

```python
soup.select('body a')# 查找body标签下面的a标签,相对路径
```

```python
soup.select('body > div')  #查找body标签下面的直接div标签
```

获取标签中的内容或属性值


```python
soup.select('body > div')[0].text         # 获得标签的中的第一条内容（字符串）
```

```python
soup.select('body > div')[1].get('class')    # 获得标签的id属性值
```

通过属性及其取值定位标签


```python
soup.select(".text")         # 通过CSS的类名查找(即class属性为text的所有标签)
```

```python
soup.select('div[class="text"]')  # 通过属性的值来查找
```

存储爬取到的网页数据

```python
title=soup.select('div[class="text"]')  # 通过属性的值来查找
for i in title:
    print(i.text)
    with open('xinwen.txt','a',encoding='utf-8') as f:               #保存为txt文档
        f.writelines(i.text+'\r')
```
